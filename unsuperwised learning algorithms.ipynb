```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("mall-customers.csv")
data.head()
```

```python
x=data.iloc[:,[3,4]].values
```

```python
s=KMeans(n_clusters=5,init="k-means++",random_state=42)
```

```python
y_pred=s.fit_predict(x) #returns clusters
y_pred
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c="blue",label="Cluster 1")
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c="green",label="Cluster 2")
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c="red",label="Cluster 3")
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=100,c="cyan",label="Cluster 4")
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=100,c="magenta",label="Cluster 5")
plt.scatter(s.cluster_centers_[:,0],s.cluster_centers_[:,1],s=200,c="yellow",label="Centroid")
plt.title("cluster of customers")
plt.xlabel("annual income")
plt.ylabel("spending score")




```

```python
k=["blue","green","red","cyan","magenta"]
l=["c1","c2","c3","c4","c5"]
for i in range(5):
    plt.scatter(x[y_pred==i,0],x[y_pred==i,1],s=50,c=k[i],label=l[i])
plt.scatter(s.cluster_centers_[:,0],s.cluster_centers_[:,1],s=200,c="yellow",label="Centroid")
plt.title("cluster of customers")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.legend()
plt.show()


```

```python
import scipy.cluster.hierarchy as shc
dendro=shc.dendrogram(shc.linkage(x,method="ward"))
plt.title("dendrogram plot")
plt.ylabel("euclidean distance")
plt.xlabel("customers")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
y_pred=hc.fit_predict(x)
```

```python
k=["blue","green","red","cyan","magenta"]
l=["c1","c2","c3","c4","c5"]
for i in range(5):
    plt.scatter(x[y_pred==i,0],x[y_pred==i,1],s=50,c=k[i],label=l[i])
#plt.scatter(s.cluster_centers_[:,0],s.cluster_centers_[:,1],s=200,c="yellow",label="Centroid")
plt.title("cluster of customers")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
